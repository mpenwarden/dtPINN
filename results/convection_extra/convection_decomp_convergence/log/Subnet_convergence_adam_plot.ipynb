```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('convection_subdomain_adam_1_log.pickle', 'rb') as handle:
    dict_1 = pickle.load(handle)

with open('convection_subdomain_adam_2_log.pickle', 'rb') as handle:
    dict_2 = pickle.load(handle)

with open('convection_subdomain_adam_4_log.pickle', 'rb') as handle:
    dict_4 = pickle.load(handle)
    
with open('convection_subdomain_adam_8_log.pickle', 'rb') as handle:
    dict_8 = pickle.load(handle)
    
with open('convection_subdomain_adam_10_log.pickle', 'rb') as handle:
    dict_10 = pickle.load(handle)
    
with open('convection_subdomain_adam_14_log.pickle', 'rb') as handle:
    dict_14 = pickle.load(handle)
    
with open('convection_subdomain_adam_16_log.pickle', 'rb') as handle:
    dict_16 = pickle.load(handle)
```

```python
cutoffs = np.linspace(0,1,17)
sub_l2_error_ls_1 = []; sub_l2_error_ls_2 = []; sub_l2_error_ls_4 = []; sub_l2_error_ls_8 = []; sub_l2_error_ls_10 = []; sub_l2_error_ls_14 = []; sub_l2_error_ls_16 = []

t_decomp = dict_1['t_decomp']
u_exact_decomp = dict_1['u_test_decomp']
u_pred_decomp = dict_1['u_pred_decomp']
for j in range(16):
    mask1 = t_decomp < cutoffs[j+1]
    mask2 = t_decomp[mask1] > cutoffs[j] 
    sub_l2_error_ls_1.append(np.linalg.norm(u_exact_decomp[mask1][mask2]-u_pred_decomp[mask1][mask2].flatten(), 2)/np.linalg.norm(u_exact_decomp[mask1][mask2], 2))
    
t_decomp = dict_2['t_decomp']
u_exact_decomp = dict_2['u_test_decomp']
u_pred_decomp = dict_2['u_pred_decomp']
for j in range(16):
    mask1 = t_decomp < cutoffs[j+1]
    mask2 = t_decomp[mask1] > cutoffs[j] 
    sub_l2_error_ls_2.append(np.linalg.norm(u_exact_decomp[mask1][mask2]-u_pred_decomp[mask1][mask2].flatten(), 2)/np.linalg.norm(u_exact_decomp[mask1][mask2], 2))
    
t_decomp = dict_4['t_decomp']
u_exact_decomp = dict_4['u_test_decomp']
u_pred_decomp = dict_4['u_pred_decomp']
for j in range(16):
    mask1 = t_decomp < cutoffs[j+1]
    mask2 = t_decomp[mask1] > cutoffs[j] 
    sub_l2_error_ls_4.append(np.linalg.norm(u_exact_decomp[mask1][mask2]-u_pred_decomp[mask1][mask2].flatten(), 2)/np.linalg.norm(u_exact_decomp[mask1][mask2], 2))
    
t_decomp = dict_8['t_decomp']
u_exact_decomp = dict_8['u_test_decomp']
u_pred_decomp = dict_8['u_pred_decomp']
for j in range(16):
    mask1 = t_decomp < cutoffs[j+1]
    mask2 = t_decomp[mask1] > cutoffs[j] 
    sub_l2_error_ls_8.append(np.linalg.norm(u_exact_decomp[mask1][mask2]-u_pred_decomp[mask1][mask2].flatten(), 2)/np.linalg.norm(u_exact_decomp[mask1][mask2], 2))
    
t_decomp = dict_10['t_decomp']
u_exact_decomp = dict_10['u_test_decomp']
u_pred_decomp = dict_10['u_pred_decomp']
for j in range(16):
    mask1 = t_decomp < cutoffs[j+1]
    mask2 = t_decomp[mask1] > cutoffs[j] 
    sub_l2_error_ls_10.append(np.linalg.norm(u_exact_decomp[mask1][mask2]-u_pred_decomp[mask1][mask2].flatten(), 2)/np.linalg.norm(u_exact_decomp[mask1][mask2], 2))
    
t_decomp = dict_14['t_decomp']
u_exact_decomp = dict_14['u_test_decomp']
u_pred_decomp = dict_14['u_pred_decomp']
for j in range(16):
    mask1 = t_decomp < cutoffs[j+1]
    mask2 = t_decomp[mask1] > cutoffs[j] 
    sub_l2_error_ls_14.append(np.linalg.norm(u_exact_decomp[mask1][mask2]-u_pred_decomp[mask1][mask2].flatten(), 2)/np.linalg.norm(u_exact_decomp[mask1][mask2], 2))
    
t_decomp = dict_16['t_decomp']
u_exact_decomp = dict_16['u_test_decomp']
u_pred_decomp = dict_16['u_pred_decomp']
for j in range(16):
    mask1 = t_decomp < cutoffs[j+1]
    mask2 = t_decomp[mask1] > cutoffs[j] 
    sub_l2_error_ls_16.append(np.linalg.norm(u_exact_decomp[mask1][mask2]-u_pred_decomp[mask1][mask2].flatten(), 2)/np.linalg.norm(u_exact_decomp[mask1][mask2], 2))
```

```python
plt.rcParams['text.usetex'] = True
plt.figure(figsize=(10,5))
plt.plot(cutoffs[1:]*16, sub_l2_error_ls_1, linewidth = 1, marker = 's')
plt.plot(cutoffs[1:]*16, sub_l2_error_ls_2, linewidth = 1, marker = 'd')
plt.plot(cutoffs[1:]*16, sub_l2_error_ls_4, linewidth = 1, marker = 'o')
plt.plot(cutoffs[1:]*16, sub_l2_error_ls_8, linewidth = 1, marker = '*')
plt.plot(cutoffs[1:]*16, sub_l2_error_ls_10, linewidth = 1, marker = '^')
plt.plot(cutoffs[1:]*16, sub_l2_error_ls_14, linewidth = 1, marker = 'p')
plt.plot(cutoffs[1:]*16, sub_l2_error_ls_16, linewidth = 1, marker = 'X')
plt.xticks(cutoffs[1:]*16)
plt.legend(['Single Domain', '2 Subdomains', '4 Subdomains', '8 Subdomains', '10 Subdomains', '14 Subdomains', '16 Subdomains'], fontsize = 14, 
           ncol = 4, loc='upper left', bbox_to_anchor=(-0.05, 1.2))
plt.grid(True, which="both", ls="-", linewidth = 0.25)
plt.xlim([0.75,16.25])
plt.yscale('log')
plt.xticks(fontsize = 20); plt.yticks(fontsize = 20);
plt.xlabel('Subdomain Number', fontsize = 20)
plt.ylabel(r'Relative $L_2$ Error', fontsize = 20);
plt.savefig('subdomain_adam_convergence', dpi = 300, bbox_inches='tight')
```
